```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import time
import glob
import datetime
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
genre_track_data = pd.read_csv('../MSD-Data/datasets/genre_track_data.csv')
lyrics_data = pd.read_csv('../MSD-Data/datasets/data_final.csv')
```

## Small Dataset: Song Hotttnesss over the years by specific genres

```python
genre_track_data_2 = genre_track_data[genre_track_data.genre_id.isin(['Pop_Rock', 'Latin', 'Rap', 'Electronic', 'Country', 'RnB'])]
genre_track_data_2 = pd.DataFrame(genre_track_data_2)
genre_track_data_2 = genre_track_data_2[(genre_track_data_2.artist_hotttnesss != 0) & (genre_track_data_2.artist_familiarity != 0) & (genre_track_data_2.song_hotttnesss != 0) & (genre_track_data_2.year != 0)]
```

```python
# Create a figure
plt.figure(figsize=(10,8))

# Create a scatterplot of... 
            # Pop_Rock
plt.scatter(genre_track_data_2['year'][genre_track_data_2['genre_id'] == 'Pop_Rock'],
            genre_track_data_2['song_hotttnesss'][genre_track_data_2['genre_id'] == 'Pop_Rock'],
            # the color
            color='#150101',
            # labelled this
            label='Pop Rock')

            # Latin
plt.scatter(genre_track_data_2['year'][genre_track_data_2['genre_id'] == 'Latin'],
            genre_track_data_2['song_hotttnesss'][genre_track_data_2['genre_id'] == 'Latin'],
            # the color
            color='#6ec59f',
            # alpha
            # labelled this
            label='Latin')

            # Rap
plt.scatter(genre_track_data_2['year'][genre_track_data_2['genre_id'] == 'Rap'],
            genre_track_data_2['song_hotttnesss'][genre_track_data_2['genre_id'] == 'Rap'],
            # the color
            color='#bd6075',
            # labelled this
            label='Rap')

            # Electronic
plt.scatter(genre_track_data_2['year'][genre_track_data_2['genre_id'] == 'Electronic'],
            genre_track_data_2['song_hotttnesss'][genre_track_data_2['genre_id'] == 'Electronic'],
            # the color
            color='#675aa3',
            # labelled this
            label='Electronic')

            # Country
plt.scatter(genre_track_data_2['year'][genre_track_data_2['genre_id'] == 'Country'],
            genre_track_data_2['song_hotttnesss'][genre_track_data_2['genre_id'] == 'Country'],
            # the color
            color='#e2c890',
            # labelled this
            label='Country')


plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Year')
plt.ylabel('Song Hotttnesss')
```

# A Focus on Rap: 2Pac vs. Biggie

```python
df_rap = lyrics_data[(lyrics_data.year != 0) & (lyrics_data.genre_id == 'Rap')]
df_rap_nob2 = df_rap[(df_rap.artist_name != '2Pac') | (df_rap.artist_name != 'The Notorious B.I.G.')]
```

```python
num_biggie = lyrics_data[lyrics_data.artist_name == 'The Notorious B.I.G.'].track_id.nunique()
print "Number of tracks by Biggie: %d" % num_biggie

num_2pac = lyrics_data[lyrics_data.artist_name == '2Pac'].track_id.nunique()
print "Number of tracks by 2Pac: %d" % num_2pac
```

```python
# Create a figure
plt.figure(figsize=(10,8))

# Create a scatterplot of,
            # rappers except biggie and 2pac
plt.scatter(df_rap_nob2['artist_familiarity'],
            df_rap_nob2['artist_hotttnesss'],
            # the color
            color='#150101',
            # labelled this
            label='Other Rappers')

            # 2Pac
plt.scatter(df_rap['artist_familiarity'][df_rap['artist_name'] == '2Pac'],
            df_rap['artist_hotttnesss'][df_rap['artist_name'] == '2Pac'],
            # the color
            color='#6ec59f',
            # alpha 
            alpha=0.7,
            # with size
            s = 124,
            # labelled this
            label='2Pac')

            # Biggie
plt.scatter(df_rap['artist_familiarity'][df_rap['artist_name'] == 'The Notorious B.I.G.'],
            df_rap['artist_hotttnesss'][df_rap['artist_name'] == 'The Notorious B.I.G.'],
            # the color
            color='#bd6075',
            # alpha 
            alpha=0.7,
            # with size
            s = 124,
            # labelled this
            label='Biggie')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim((0,1))
plt.xlabel('Artist Familiarity')
plt.ylabel('Artist Hotttnesss')
```
